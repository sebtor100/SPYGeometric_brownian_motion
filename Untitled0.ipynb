```python
import yfinance as yf
import numpy as np
import pandas as pd
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader, TensorDataset
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error, mean_absolute_error
import warnings
warnings.filterwarnings('ignore')

class EnhancedStockPredictor(nn.Module):
    """Hybrid LSTM model that predicts price, drift, and volatility."""
    def __init__(self, input_size, hidden_size=128, num_layers=2, dropout=0.2):
        super(EnhancedStockPredictor, self).__init__()
        self.lstm = nn.LSTM(input_size=input_size, hidden_size=hidden_size, num_layers=num_layers, dropout=dropout, batch_first=True)
        self.attention = nn.MultiheadAttention(embed_dim=hidden_size, num_heads=8, dropout=dropout, batch_first=True)
        self.feature_extractor = nn.Sequential(nn.Linear(hidden_size, hidden_size // 2), nn.ReLU(), nn.Dropout(dropout), nn.Linear(hidden_size // 2, hidden_size // 4), nn.ReLU(), nn.Dropout(dropout))
        self.price_predictor = nn.Linear(hidden_size // 4, 1)
        self.volatility_predictor = nn.Linear(hidden_size // 4, 1)
        self.drift_predictor = nn.Linear(hidden_size // 4, 1)
        self.sigmoid, self.tanh = nn.Sigmoid(), nn.Tanh()

    def forward(self, x):
        lstm_out, _ = self.lstm(x)
        attn_out, _ = self.attention(lstm_out, lstm_out, lstm_out)
        features = self.feature_extractor(attn_out[:, -1, :])
        price_pred = self.price_predictor(features)
        volatility_pred = self.sigmoid(self.volatility_predictor(features)) * 0.05
        drift_pred = self.tanh(self.drift_predictor(features)) * 0.03
        return price_pred, volatility_pred, drift_pred

def fetch_and_clean_data(ticker, period='10y'):
    """Fetch stock data and ensure clean column structure."""
    data = yf.download(ticker, period=period, auto_adjust=True)
    if isinstance(data.columns, pd.MultiIndex):
        data.columns = data.columns.get_level_values(0)
    return data.loc[:, ~data.columns.duplicated()].dropna()

def create_enhanced_features(stock_data):
    """Create a comprehensive feature set with robust data handling."""
    df = stock_data.copy()
    for col in ['Open', 'High', 'Low', 'Close', 'Volume']:
        if col in df.columns and isinstance(df[col], pd.DataFrame):
            df[col] = df[col].iloc[:, 0]

    df['Log_Returns'] = np.log(df['Close'] / df['Close'].shift(1))
    for window in [5, 10, 20, 50]:
        df[f'MA_{window}'] = df['Close'].rolling(window=window).mean()
        df[f'MA_Ratio_{window}'] = df['Close'] / df[f'MA_{window}']
    df['Volatility_20'] = df['Log_Returns'].rolling(window=20).std()
    delta = df['Close'].diff(); gain = (delta.where(delta > 0, 0)).rolling(window=14).mean(); loss = (-delta.where(delta < 0, 0)).rolling(window=14).mean()
    df['RSI'] = 100 - (100 / (1 + (gain / loss))); exp1 = df['Close'].ewm(span=12, adjust=False).mean(); exp2 = df['Close'].ewm(span=26, adjust=False).mean()
    df['MACD'] = exp1 - exp2; df['MACD_Signal'] = df['MACD'].ewm(span=9, adjust=False).mean()
    feature_columns = [col for col in df.columns if col not in ['Open', 'High', 'Low', 'Close', 'Volume']]
    return df.dropna(), feature_columns

def prepare_sequences_multi_target(data, feature_columns, sequence_length=60):
    """Prepare sequences with multiple targets (price, drift, vol)."""
    sequences, targets_price, targets_drift, targets_vol = [], [], [], []
    data['Hist_Drift'] = data['Log_Returns'].rolling(window=sequence_length).mean()
    data['Hist_Vol'] = data['Log_Returns'].rolling(window=sequence_length).std()
    data = data.dropna()
    for i in range(len(data) - sequence_length):
        sequences.append(data[feature_columns].iloc[i:i+sequence_length].values)
        target_idx = i + sequence_length
        targets_price.append(data['Close'].iloc[target_idx]); targets_drift.append(data['Hist_Drift'].iloc[target_idx]); targets_vol.append(data['Hist_Vol'].iloc[target_idx])
    return np.array(sequences), np.array(targets_price), np.array(targets_drift), np.array(targets_vol)

# --- CHANGE: Upgraded simulation to handle a time-varying drift path ---
def enhanced_gbm_simulation(S0, mu, sigma, T, N, num_simulations=2000):
    """Simulate price paths using Geometric Brownian Motion with time-varying drift."""
    dt = T / N
    np.random.seed(42)
    dW = np.random.normal(0, np.sqrt(dt), (num_simulations, N))
    paths = np.zeros((num_simulations, N + 1))
    paths[:, 0] = S0

    # Check if mu is a single value or a path
    is_mu_path = isinstance(mu, (list, np.ndarray))

    for t in range(1, N + 1):
        current_mu = mu[t-1] if is_mu_path else mu
        drift = (current_mu - 0.5 * sigma**2) * dt
        diffusion = sigma * dW[:, t-1]
        paths[:, t] = paths[:, t-1] * np.exp(drift + diffusion)
    return np.linspace(0, T, N + 1), paths.T

def plot_training_results(metrics, ticker):
    """Plot model training and evaluation metrics."""
    fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(16, 12)); fig.suptitle(f'{ticker} AI Model Evaluation', fontsize=16, fontweight='bold')
    ax1.plot(metrics['train_losses'], color='blue', linewidth=2); ax1.set_title('Training Loss (MSE) Over Epochs'); ax1.set_xlabel('Epoch'); ax1.set_ylabel('Total MSE Loss'); ax1.grid(True, linestyle='--', linewidth=0.5)
    ax2.scatter(metrics['test_actual'], metrics['test_pred'], alpha=0.6, color='red', edgecolor='k', s=50); min_val, max_val = min(metrics['test_actual'].min(), metrics['test_pred'].min()), max(metrics['test_actual'].max(), metrics['test_pred'].max()); ax2.plot([min_val, max_val], [min_val, max_val], 'k--', linewidth=2, label='Perfect Prediction'); ax2.set_xlabel('Actual Prices ($)'); ax2.set_ylabel('Predicted Prices ($)'); ax2.set_title('Test Set: Actual vs. Predicted Prices'); ax2.legend(); ax2.grid(True, linestyle='--', linewidth=0.5)
    errors = metrics['test_pred'] - metrics['test_actual']; ax3.hist(errors, bins=50, alpha=0.75, color='green', edgecolor='black'); ax3.axvline(0, color='red', linestyle='--', linewidth=2, label='Zero Error'); ax3.set_xlabel('Prediction Error ($)'); ax3.set_ylabel('Frequency'); ax3.set_title('Distribution of Prediction Errors'); ax3.legend(); ax3.grid(True, linestyle='--', linewidth=0.5)
    ax4.scatter(range(len(metrics['test_vol'])), metrics['test_vol'], alpha=0.6, color='orange', label='Volatility', s=20); ax4_twin = ax4.twinx(); ax4_twin.scatter(range(len(metrics['test_drift'])), metrics['test_drift'], alpha=0.6, color='purple', label='Drift', s=20); ax4.set_xlabel('Test Sample Index'); ax4.set_ylabel('Predicted Volatility', color='orange'); ax4_twin.set_ylabel('Predicted Drift', color='purple'); ax4.set_title('Predicted Parameters on Test Set'); ax4.tick_params(axis='y', labelcolor='orange'); ax4_twin.tick_params(axis='y', labelcolor='purple'); ax4.grid(True, linestyle='--', linewidth=0.5)
    plt.tight_layout(rect=[0, 0.03, 1, 0.95]); plt.show()


def enhanced_analysis_and_visualization(
    ticker, model, scaler_X, scaler_y,
    enhanced_data, feature_columns,
    forecast_months=6, sequence_length=60, smoothing_window=30,
    alpha=0.25
):
    """Perform enhanced analysis with a blended, mean-reverting forecast model."""
    print(f"\nðŸ”® AI-Powered Advanced Analysis for {ticker}"); print("="*60)
    current_price = enhanced_data['Close'].iloc[-1]

    # --- Smoothing Logic ---
    recent_drift_preds, recent_vol_preds = [], []
    start_idx = len(enhanced_data) - (smoothing_window + sequence_length)
    smoothing_data = enhanced_data.iloc[start_idx:]
    device = torch.device('cuda' if torch.cuda.is_available() else 'cpu'); model.to(device); model.eval()
    with torch.no_grad():
        for i in range(smoothing_window):
            sequence_data = smoothing_data[feature_columns].iloc[i:i+sequence_length].values
            sequence_tensor = torch.FloatTensor(scaler_X.transform(sequence_data)).unsqueeze(0).to(device)
            _, vol_pred, drift_pred = model(sequence_tensor)
            recent_vol_preds.append(vol_pred.cpu().item()); recent_drift_preds.append(drift_pred.cpu().item())

    ai_drift = np.mean(recent_drift_preds) * 252; ai_vol = np.mean(recent_vol_preds) * np.sqrt(252)
    trad_drift = enhanced_data['Log_Returns'].mean() * 252; trad_vol = enhanced_data['Log_Returns'].std() * np.sqrt(252)

    # --- CHANGE: Create a time-varying drift path for mean reversion ---
    initial_blended_drift = trad_drift + alpha * (ai_drift - trad_drift)
    T, N = forecast_months / 12, forecast_months * 21
    # Create a path that linearly interpolates from the initial blended drift to the final traditional drift
    drift_path = np.linspace(initial_blended_drift, trad_drift, N)

    print(f"Current Price: ${current_price:.2f}")
    print(f"Long-Term Traditional Drift: {trad_drift*100:.2f}%")
    print(f"Initial Blended Drift: {initial_blended_drift*100:.2f}% (reverts to {trad_drift*100:.2f}%)")

    # Run simulations with the new time-varying drift path
    _, blended_paths = enhanced_gbm_simulation(current_price, drift_path, ai_vol, T, N)
    _, trad_paths = enhanced_gbm_simulation(current_price, trad_drift, trad_vol, T, N)

    fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(18, 14), gridspec_kw={'height_ratios': [1, 1.2]})
    fig.suptitle(f'{ticker} - Forecast Analysis: Mean-Reverting AI vs. Traditional ({forecast_months} Months)', fontsize=18, fontweight='bold')

    # Plotting code is now more meaningful as it reflects the new forecast logic
    historical_prices = enhanced_data['Close'].iloc[-252:]; ax1.plot(historical_prices.index, historical_prices, label='Historical Price (1 Year)', color='navy', linewidth=2); ax1.axhline(y=current_price, color='red', linestyle='--', alpha=0.8, label=f'Current: ${current_price:.2f}'); ax1.set_title('Recent Price History', fontsize=14); ax1.set_ylabel('Price ($)'); ax1.grid(True, linestyle='--', linewidth=0.5); ax1.legend(); fig.autofmt_xdate()
    days = np.arange(N + 1); ax2.plot(days, np.mean(blended_paths, axis=1), color='red', linewidth=2.5, label='Mean-Reverting AI (Mean)'); ax2.fill_between(days, np.percentile(blended_paths, 5, axis=1), np.percentile(blended_paths, 95, axis=1), alpha=0.2, color='red', label='AI 90% CI'); ax2.plot(days, np.mean(trad_paths, axis=1), color='blue', linewidth=2.5, linestyle='--', label='Traditional GBM (Mean)'); ax2.fill_between(days, np.percentile(trad_paths, 5, axis=1), np.percentile(trad_paths, 95, axis=1), alpha=0.2, color='blue', label='Traditional 90% CI'); ax2.set_title('Forecast Comparison', fontsize=14); ax2.set_ylabel('Predicted Price ($)'); ax2.set_xlabel('Trading Days from Now'); ax2.legend(); ax2.grid(True, linestyle='--', linewidth=0.5)
    blended_final_prices, trad_final_prices = blended_paths[-1, :], trad_paths[-1, :]; ax3.hist(blended_final_prices, bins=60, alpha=0.7, color='red', label='Mean-Reverting AI', density=True); ax3.set_xlabel('Price ($)'); ax3.set_ylabel('AI Density', color='red'); ax3.tick_params(axis='y', labelcolor='red'); ax3_twin = ax3.twinx(); ax3_twin.hist(trad_final_prices, bins=60, alpha=0.7, color='blue', label='Traditional', density=True); ax3_twin.set_ylabel('Traditional Density', color='blue'); ax3_twin.tick_params(axis='y', labelcolor='blue'); ax3.axvline(current_price, color='black', linestyle='-', linewidth=2, label=f'Current: ${current_price:.2f}'); ax3.axvline(np.mean(blended_final_prices), color='darkred', linestyle='--', linewidth=2, label=f'AI Mean: ${np.mean(blended_final_prices):.2f}'); ax3.axvline(np.mean(trad_final_prices), color='darkblue', linestyle='--', linewidth=2, label=f'Trad. Mean: ${np.mean(trad_final_prices):.2f}'); ax3.set_title('Final Price Distribution', fontsize=14); lines, labels = ax3.get_legend_handles_labels(); lines2, labels2 = ax3_twin.get_legend_handles_labels(); ax3_twin.legend(lines + lines2, labels + labels2, loc=0); ax3.grid(True, linestyle='--', linewidth=0.5, axis='x')
    percentiles = [5, 25, 50, 75, 95]; blended_percentiles, trad_percentiles = np.percentile(blended_final_prices, percentiles), np.percentile(trad_final_prices, percentiles); x_pos = np.arange(len(percentiles)); width = 0.35; ax4.bar(x_pos - width/2, blended_percentiles, width, label='Mean-Reverting AI', color='red', alpha=0.8); ax4.bar(x_pos + width/2, trad_percentiles, width, label='Traditional', color='blue', alpha=0.8); ax4.axhline(y=current_price, color='black', linestyle='-', linewidth=2, label='Current Price'); ax4.set_title('Risk Percentile Comparison', fontsize=14); ax4.set_ylabel('Price ($)'); ax4.set_xticks(x_pos); ax4.set_xticklabels([f'{p}th' for p in percentiles]); ax4.legend(); ax4.grid(axis='y', linestyle='--', linewidth=0.5)
    plt.tight_layout(rect=[0, 0.03, 1, 0.95]); plt.show()


def train_enhanced_model(ticker, sequence_length=60, epochs=100):
    """Train the enhanced AI model with a multi-task loss function."""
    print(f"ðŸš€ Training Enhanced AI Model for {ticker}"); print("="*60)
    stock_data = fetch_and_clean_data(ticker, period='10y')
    enhanced_data, feature_columns = create_enhanced_features(stock_data)
    print(f"Using {len(feature_columns)} features.")
    sequences, targets_price, targets_drift, targets_vol = prepare_sequences_multi_target(enhanced_data, feature_columns, sequence_length)
    split_idx = int(0.8 * len(sequences)); X_train, X_test = sequences[:split_idx], sequences[split_idx:]; y_train_price, y_test_price = targets_price[:split_idx], targets_price[split_idx:]; y_train_drift, _ = targets_drift[:split_idx], targets_drift[split_idx:]; y_train_vol, _ = targets_vol[:split_idx], targets_vol[split_idx:]; scaler_X, scaler_y = MinMaxScaler(), MinMaxScaler(); X_train_reshaped = X_train.reshape(-1, X_train.shape[-1]); scaler_X.fit(X_train_reshaped); X_train_scaled = scaler_X.transform(X_train_reshaped).reshape(X_train.shape); X_test_scaled = scaler_X.transform(X_test.reshape(-1, X_test.shape[-1])).reshape(X_test.shape); y_train_price_scaled = scaler_y.fit_transform(y_train_price.reshape(-1, 1)).flatten(); X_train_tensor, X_test_tensor = torch.FloatTensor(X_train_scaled), torch.FloatTensor(X_test_scaled); y_train_tensor_price, y_train_tensor_drift, y_train_tensor_vol = torch.FloatTensor(y_train_price_scaled), torch.FloatTensor(y_train_drift), torch.FloatTensor(y_train_vol)
    train_dataset = TensorDataset(X_train_tensor, y_train_tensor_price, y_train_tensor_drift, y_train_tensor_vol); train_loader = DataLoader(train_dataset, batch_size=64, shuffle=True)
    device = torch.device('cuda' if torch.cuda.is_available() else 'cpu'); print(f"Using device: {device}")
    model = EnhancedStockPredictor(input_size=len(feature_columns)).to(device)
    criterion_price, criterion_drift, criterion_vol = nn.MSELoss(), nn.MSELoss(), nn.MSELoss()
    optimizer = optim.Adam(model.parameters(), lr=0.001); scheduler = optim.lr_scheduler.ReduceLROnPlateau(optimizer, 'min', patience=10, factor=0.5)
    w_price, w_drift, w_vol = 1.0, 0.5, 0.5; train_losses = []
    for epoch in range(epochs):
        model.train(); epoch_loss = 0
        for batch_X, batch_y_price, batch_y_drift, batch_y_vol in train_loader:
            batch_X, batch_y_price, batch_y_drift, batch_y_vol = batch_X.to(device), batch_y_price.to(device), batch_y_drift.to(device), batch_y_vol.to(device)
            optimizer.zero_grad(); pred_price_scaled, pred_drift, pred_vol = model(batch_X)
            loss_p = criterion_price(pred_price_scaled.squeeze(), batch_y_price); loss_d = criterion_drift(pred_drift.squeeze(), batch_y_drift); loss_v = criterion_vol(pred_vol.squeeze(), batch_y_vol)
            total_loss = w_price * loss_p + w_drift * loss_d + w_vol * loss_v
            total_loss.backward(); torch.nn.utils.clip_grad_norm_(model.parameters(), 1.0); optimizer.step(); epoch_loss += total_loss.item()
        avg_loss = epoch_loss / len(train_loader); train_losses.append(avg_loss); scheduler.step(avg_loss)
        if (epoch + 1) % 10 == 0: print(f"Epoch {epoch+1}/{epochs}, Loss: {avg_loss:.6f}, LR: {optimizer.param_groups[0]['lr']:.6f}")
    model.eval()
    with torch.no_grad():
        test_pred_scaled, test_vol, test_drift = model(X_test_tensor.to(device)); test_pred_actual = scaler_y.inverse_transform(test_pred_scaled.cpu().numpy())
    mse = mean_squared_error(y_test_price, test_pred_actual)
    print(f"\nðŸ“Š Model Performance (Test Set):"); print(f"MSE: {mse:.4f}, MAE: {mean_absolute_error(y_test_price, test_pred_actual):.4f}, RMSE: {np.sqrt(mse):.4f}")
    metrics = {'test_pred': test_pred_actual.flatten(), 'test_actual': y_test_price.flatten(), 'test_vol': test_vol.cpu().numpy().flatten(), 'test_drift': test_drift.cpu().numpy().flatten(), 'train_losses': train_losses}
    plot_training_results(metrics, ticker)
    return model, scaler_X, scaler_y, enhanced_data, feature_columns

# MAIN EXECUTION BLOCK
if __name__ == "__main__":
    ticker = "SPY"
    try:
        model, scaler_X, scaler_y, enhanced_data, feature_columns = train_enhanced_model(ticker=ticker, epochs=100)
        print("\nâœ… Model training completed successfully!")
        enhanced_analysis_and_visualization(
            ticker=ticker, model=model, scaler_X=scaler_X, scaler_y=scaler_y,
            enhanced_data=enhanced_data, feature_columns=feature_columns,
            forecast_months=6, alpha=0.25 # You can tune this alpha (0.0 to 1.0) to control AI influence
        )
    except Exception as e:
        print(f"\nâŒ An error occurred during execution: {str(e)}"); import traceback; traceback.print_exc()
```
